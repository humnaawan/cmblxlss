Takahashi halo catalog file: skyhalo_nres12r000.halo: 40GB file.

Cant be read in locally, so access chunks until have a specified final galaxy sample size (with pid and mass cuts).

```python
import numpy as np
import pandas as pd
import os
```

```python
os.chdir('..')
home= os.getcwd()
print(home)
```

```python
haloCatPath= '/Volumes/My_Passport_Mac_Current/skyhalo_nres12r000.halo'
```

```python
outDir= '%s/data/interm/'%home
if not os.path.exists(outDir): os.makedirs(outDir)
print(outDir)
```

```python
def read_Takahashi_HaloCat(path, nrows= 30, skiprows= 0):
    # skiprows input should not include the header
    header= pd.read_table(path, nrows=4, header=None, )
    data= pd.read_table(path, nrows= nrows, header=None, skiprows= 4+skiprows)
    
    head= [(f) for f in header[0][0].split(' ') if (f!='' and not f.__contains__('#') and not f.__contains__('rock'))]
    nCol= len(head)
    
    
    for i, row in enumerate(data[0][:]):
        if (i==0): 
            dataArr= np.array([float(f) for f in row.split(' ') if (f!='' and not f.__contains__('list'))])
        else:
            dataArr= np.vstack([dataArr, np.array([float(f) for f in row.split(' ') if (f!='' and not f.__contains__('list'))])])

    print('%s rows read in.'%len(dataArr))
    print('%s columns:\n%s\n'%(nCol, head))
    return pd.DataFrame(dataArr, columns= head)
```

```python
goodN= 0

i= 0
chunkSize= 10000

keep= ['ID', 'Mvir', 'z_halo', 'theta_i', 'phi_i']#, 'theta_s', 'phi_s', 'ipix', 'kplane+1']

Mmin, Mmax= 5e12, 8e13
while (goodN<10000): #00):
    data= read_Takahashi_HaloCat(path= haloCatPath, nrows= chunkSize, skiprows= i*chunkSize)
    
    # find all the galaxies with parentID= -1 (central gals).
    # sample cuts: Mvir between 5e12 to 8e13 for now.
    subData= data[data.parent_ID == -1]
    subData.reset_index(drop=True)
    nKeep= len(subData)
    print('Dropped %s rows with parentID!= -1'%(len(data)-nKeep) )
    print('Final nRows: %s\n'%nKeep)

    # implement mass cuts
    subData= subData[(subData.Mvir> Mmin) & (subData.Mvir< Mmax)]
    subData.reset_index(drop=True)

    print('Dropped %s rows with Mvir not between 5e12 to 8e13'%(nKeep-len(subData)))
    print('Final nRows: %s'%len(subData))

    goodN+= len(subData)
    
    filename= '%s/output_%s.csv'%(outDir, i)
    subData.to_csv(filename, columns= keep)

    print('\n## Saved %s\n'%filename)
    i+=1
    
    print('%s galaxies; iteration # %s\n'%(goodN, i))
    print('------------------------------')
```

```python
# combine the files now.
os.chdir(outDir)
filenames= [f for f in os.listdir(outDir) if not f.__contains__('gals')]

result= pd.read_csv(filenames[0])

for filename in filenames[1:]:
    print('Reading %s'%filename)
    result = pd.concat([result, pd.read_csv(filename)], ignore_index=True)

filename= '%s/finalOutput_%sgals.csv'%(outDir, len(result))
result.to_csv(filename, columns= keep, index=False)

print('\n## Saved %s\n'%filename)
```
